## 실습 3

1. UniversalBank.csv를 sub라는 데이터프레임으로 불러오기
2. sub에서 가족 수(Family)가 1명인 데이터만 sub1라는 새로운 데이터프레임으로 만들기
3. sub1 데이터에서 'Experience'와 'Income'의 중간값 구하기
4. sub에서 가족 수(Family)가 3명인 데이터만 sub2라는 새로운 데이터프레임으로 만들기
5. 가족 수(Family)가 1명이거나 3명’인 데이터프레임 구하기

```python
#1. UniversalBank.csv를 sub라는 데이터프레임으로 불러오기
filepath1 = '/content/drive/MyDrive/비타민_11기_정규세션/230225/UniversalBank.csv'
sub = pd.read_csv(filepath1)
```

```python
#2. sub에서 가족 수(Family)가 1명인 데이터만 sub1라는 새로운 데이터프레임으로 만들기
sub1 = sub[sub.Family == 1]
sub1
```

```python
#3. sub1 데이터에서 'Experience'와 'Income'의 중간값 구하기
sub1 = sub.copy()
exp_mid = sub1['Experience'].median()
inc_mid = sub1['Income'].median()
print('Experience 중간값 : ', exp_mid)
print('Income 중간값 : ', inc_mid)
```

```python
#4. sub에서 가족 수(Family)가 3명인 데이터만 sub2라는 새로운 데이터프레임으로 만들기
sub2 = sub[sub['Family'] == 3]
sub2
```

```python
#5. 가족 수(Family)가 1명이거나 3명인 데이터프레임 구하기 (sub1와 sub2를 세로방향으로 합치기)
sub3 = pd.concat([sub1, sub2], axis = 0)
sub3
```

## 실습 4

1. AustralianWines.csv를 wine이라는 데이터프레임으로 불러오기
2. wine에서 첫 10행만 보기
3. wine에서 정수형 변수의 개수
4. wine에서 Rose의 판매 개수가 150개 이상인 행을 wine_r 이라는 데이터프레임으로 저장하기
5. wine_r의 변수를 Month, Red, Rose, Sparkling 만 남기기
6. wine_r의 열(변수)이름을 날짜, 레드, 로제, 스파클링으로 변경하기

```python
#1. AustralianWines.csv를 wine이라는 데이터프레임으로 불러오기
filepath2 = '/content/drive/MyDrive/비타민_11기_정규세션/230225/AustralianWines.csv'
wine = pd.read_csv(filepath2)
```

```python
#2. wine에서 첫 10행만 보기
wine.head(10)
```

```python
#3. wine에서 정수형 변수의 개수는?
wine.count()
```

```python
#4. wine에서 Rose의 판매 개수가 150개 이상인 행을 wine_r 이라는 데이터프레임으로 저장하기
wine_r = wine[wine['Rose'] >= 150]
wine_r
```

```python
#5. wine_r의 변수를 Month, Red, Rose, Sparkling 만 남기기
wine_r = wine_r[['Month', 'Red', 'Rose', 'Sparkling']]
wine_r
```

```python
#6. wine_r의 열(변수)이름을 '날짜', '레드','로제', '스파클링'으로 변경하기
wine_r = wine_r.rename(columns = {'Month' : '날짜', 'Red' : '레드', 'Rose' : '로제', 'Sparkling' : '스파클링'})
wine_r
```
